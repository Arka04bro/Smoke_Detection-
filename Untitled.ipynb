```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

import torch
import cv2
import telegram
import asyncio
import logging
import os
from dotenv import load_dotenv
import time
import requests
from io import BytesIO
import numpy as np
'''
Code was written by Eraly Gainulla 15.01.2025 
Updated for ESP32-CAM integration
'''

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö c –æ–∫—Ä—É–∂–µ–Ω–∏—è
load_dotenv(dotenv_path="/home/eraly/projects/ml_dl/.env")

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ª–æ–≥–∏—Ä–æ–≤–∞–Ω–∏—è
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filename="smoke_detection.log"
)
logger = logging.getLogger(__name__)

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏–∏
MODEL_PATH = os.getenv("MODEL_PATH", "bestyolo.pt")
BOT_TOKEN = os.getenv('BOT_TOKEN', '7680528919:AAGmb6gQJVJhVT8inRDCnbV7razhA_RYe4g')
CHAT_ID = os.getenv('CHAT_ID', '-4670568334')
CONFIDENCE_THRESHOLD = float(os.getenv('CONFIDENCE_THRESHOLD', 0.1))
CONSECUTIVE_FRAMES_THRESHOLD = int(os.getenv('CONSECUTIVE_FRAMES_THRESHOLD', 10))
USE_GPU = os.getenv("USE_GPU", "True").lower() in ["true", "1", "yes"]
ESP32CAM_IP = os.getenv("ESP32CAM_IP", "192.168.100.252")
ESP32CAM_STREAM_URL = f"http://{ESP32CAM_IP}/capture"  # URL –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –∫–∞–¥—Ä–æ–≤ —Å ESP32-CAM

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π –≤ Telegram 
async def send_telegram_notification(bot, chat_id, frame_path):
    try:
        await bot.send_message(chat_id=chat_id, text="–û–±–Ω–∞—Ä—É–∂–µ–Ω –¥—ã–º!")
        with open(frame_path, "rb") as photo:
            await bot.send_photo(chat_id=chat_id, photo=photo)
        logger.info("–£–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–æ –≤ Telegram.")
    except Exception as e:
        logger.error(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –æ—Ç–ø—Ä–∞–≤–∫–µ —É–≤–µ–¥–æ–º–ª–µ–Ω–∏—è: {e}")

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ –º–æ–¥–µ–ª–∏
def load_model():
    try:
        model = torch.hub.load("ultralytics/yolov5", "custom", path='bestyolo.pt', force_reload=True)
        device = "cuda" if USE_GPU and torch.cuda.is_available() else "cpu"
        model.to(device)
        logger.info(f"–ú–æ–¥–µ–ª—å —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –Ω–∞ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {device}")
        return model, device
    except Exception as e:
        logger.error(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–≥—Ä—É–∑–∫–µ –º–æ–¥–µ–ª–∏: {e}")
        raise

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –∫–∞–¥—Ä–∞ —Å ESP32-CAM
def get_esp32cam_frame():
    try:
        response = requests.get(ESP32CAM_STREAM_URL, timeout=5)
        if response.status_code == 200:
            img_array = np.asarray(bytearray(response.content), dtype=np.uint8)
            frame = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
            return True, frame
        else:
            logger.error(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –ø–æ–ª—É—á–µ–Ω–∏–∏ –∫–∞–¥—Ä–∞: HTTP {response.status_code}")
            return False, None
    except Exception as e:
        logger.error(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –ø–æ–¥–∫–ª—é—á–µ–Ω–∏–∏ –∫ ESP32-CAM: {e}")
        return False, None

# –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
async def main():
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Telegram-–±–æ—Ç–∞
    if not BOT_TOKEN or not CHAT_ID:
        logger.error("–¢–æ–∫–µ–Ω –±–æ—Ç–∞ –∏/–∏–ª–∏ chat_id –Ω–µ —É–∫–∞–∑–∞–Ω—ã. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏.")
        return
    bot = telegram.Bot(token=BOT_TOKEN)

    # –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏
    model, device = load_model()

    # –ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –¥–ª—è –∫–æ–Ω—Ç—Ä–æ–ª—è —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π
    smoke_detected = False
    frames_without_smoke = 0
    last_detection_time = 0  # –í—Ä–µ–º—è –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –¥—ã–º–∞

    try:
        while True:
            # –ü–æ–ª—É—á–∞–µ–º –∫–∞–¥—Ä —Å ESP32-CAM
            ret, frame = get_esp32cam_frame()
            if not ret:
                logger.error("–ù–µ —É–¥–∞–ª–æ—Å—å –ø–æ–ª—É—á–∏—Ç—å –∫–∞–¥—Ä —Å ESP32-CAM. –ü–æ–≤—Ç–æ—Ä–Ω–∞—è –ø–æ–ø—ã—Ç–∫–∞ —á–µ—Ä–µ–∑ 5 —Å–µ–∫—É–Ω–¥...")
                await asyncio.sleep(5)
                continue

            # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –∏–Ω—Ñ–µ—Ä–µ–Ω—Å–∞
            results = model(frame)
            detections = results.xyxy[0].cpu().numpy()

            smoke_present = False
            for det in detections:
                class_id = int(det[5])
                confidence = det[4]
                if class_id == 0 and confidence > CONFIDENCE_THRESHOLD:
                    smoke_present = True
                    x1, y1, x2, y2 = map(int, det[:4])
                    cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
                    label = f"Smoke: {confidence:.2f}"
                    cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

            # –û–±—Ä–∞–±–æ—Ç–∫–∞ —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π
            if smoke_present and not smoke_detected:
                current_time = time.time()
                if current_time - last_detection_time >= 10:  # –¢–∞–π–º-–∞—É—Ç 10 —Å–µ–∫—É–Ω–¥
                    frame_path = "smoke_detected.jpg"
                    cv2.imwrite(frame_path, frame)
                    await send_telegram_notification(bot, CHAT_ID, frame_path)
                    smoke_detected = True
                    frames_without_smoke = 0
                    last_detection_time = current_time

            if not smoke_present:
                frames_without_smoke += 1
                if smoke_detected and frames_without_smoke >= CONSECUTIVE_FRAMES_THRESHOLD:
                    smoke_detected = False

            # –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –∫–∞–¥—Ä–∞ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
            resized_frame = cv2.resize(frame, (1280, 720))

            # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∫–∞–¥—Ä–∞ —Å –∏–∑–º–µ–Ω–µ–Ω–Ω—ã–º —Ä–∞–∑–º–µ—Ä–æ–º
            cv2.imshow("Smoke Detection - ESP32-CAM", resized_frame)
            if cv2.waitKey(1) == ord("q"):
                break

            await asyncio.sleep(0.1)  # –ù–µ–±–æ–ª—å—à–∞—è –∑–∞–¥–µ—Ä–∂–∫–∞ –¥–ª—è —Å–Ω–∏–∂–µ–Ω–∏—è –Ω–∞–≥—Ä—É–∑–∫–∏

    except Exception as e:
        logger.error(f"–û—à–∏–±–∫–∞ –≤–æ –≤—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è: {e}")
    finally:
        cv2.destroyAllWindows()
        logger.info("–ü—Ä–æ–≥—Ä–∞–º–º–∞ –∑–∞–≤–µ—Ä—à–µ–Ω–∞.")

await main()
```
